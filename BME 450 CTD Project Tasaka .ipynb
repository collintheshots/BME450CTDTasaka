## Plotting


```python
import matplotlib.pyplot as plt
```

```python
# Temperature
T = fileName.seawater_temperature
```

```python
# Pressure
P = fileName.seawater_pressure
```

```python
plt.plot(T,P)
plt.xlabel('Temperature')
plt.ylabel('Pressure')
plt.show()
```

```python
# practical salinity
S = fileName.practical_salinity

# depth: 1 meter sea water = 1 dbar. P is given in dbar. Z represents P in meters sea water (msw)
Z = P
```

```python
# T^2 = T*T
T2=T*T

# T^3 =T*T*T
T3=T*T*T
```

```python
# c = speed of sound

c = 1449.2 + (4.6 * T) - (.055 * T2) + (.00029 * T3) + ((1.34 - 0.01*T)*(S - 35)) + (.016*Z)
```

```python
plt.plot(c,Z)
plt.xlabel('Speed of Sound')
plt.ylabel('Depth')
plt.title('Depth vs Speed of Sound')
plt.show()
```

```python
# time
t = fileName.time
```

```python
plt.plot(t,Z)
plt.xlabel('time')
plt.ylabel('depth')
plt.title('Depth vs Time')
plt.show()
```
